```python
import pandas as pd

df = pd.read_csv("bert_errors.csv")
```

```python
print("BERT Error Word Counts")
df["sentence"].map(lambda x: len(x.split())).hist(bins=20)
```

```python
DECENT_LENGTH = 140
df["sentence"].map(lambda x: len(x) < DECENT_LENGTH).value_counts()
```

```python
print("BERT Error Character Counts")
df["sentence"].map(lambda x: len(x)).hist(bins=20)
```
